Se o numero não for igual a -1, continue.

```python
largest_number = 999999999

# insira o primeiro valor.
number = int(input("Digite um número ou valor -1 para parar: "))
# se o numero não for igual a -1, continue.
while number != -1:
  # o numero for maior que o largest_number?
  if number > largest_number:
    # sim, atualize o maior número
    largest_number = number
  # insira o proximo número.
  number = int(input("Digite um número ou valor -1 para parar: "))
print("O maior numero é: ", largest_number)
```
