```python
aluno = {}
aluno['nome'] = input('Digite o nome do aluno: ')
aluno['media'] = float(input(f'Digite a media do aluno: {aluno["nome"]}: '))
if aluno['media'] >= 7:
  aluno['situação'] = 'Aprovado'
elif 5 <= aluno['media'] < 7:
  aluno['situação'] = 'Recuperação'
else:
  aluno['situação'] = 'Reprovado'
print('=='*30)
for k, v in aluno.items():
  print(f' - {k} é igual a {v}')
```

```python
lista_pessoas_camisas = []

for i in range(2):
    print(f"\n--- Inserindo informações da pessoa {i+1} ---")
    nome_pessoa = input('Digite o nome da pessoa: ')
    cor_camisa = input(f'Digite a cor da camisa de {nome_pessoa}: ')

    info_pessoa = {
        'Nome': nome_pessoa,
        'Cor da Camisa': cor_camisa
    }
    lista_pessoas_camisas.append(info_pessoa)

print('\n--- Todas as informações inseridas ---')
for pessoa in lista_pessoas_camisas:
    print(pessoa)
```

```python
semana = {}

semana['Segunda-feira'] = ['Matemática', 'Português', 'História']
semana['Terça-feira'] = ['Geografia', 'Ciências', 'Inglês']
semana['Quarta-feira'] = ['Matemática', 'Educação Física', 'Artes']
semana['Quinta-feira'] = ['Português', 'Física', 'Química']
semana['Sexta-feira'] = ['Biologia', 'Filosofia', 'Sociologia']
semana['Sábado'] = [] # Sem aulas
semana['Domingo'] = [] # Sem aulas

print("\n --- Dicionário 'semana' com as aulas por dia: ---")
for dia, aulas in semana.items(): # dia será a chave, aulas será o valor -> é uma tupla
    if aulas:
        print(f'{dia}: {", ".join(aulas)}')
    else:
        print(f'{dia}: Nenhuma aula')
```

```python
dia_desejado = input('Qual dia da semana você quer consultar? (Ex: Segunda-feira): ')

dia_formatado = dia_desejado.capitalize()

if dia_formatado in semana:
    aulas_do_dia = semana[dia_formatado]
    if aulas_do_dia:
        print(f'Aulas de {dia_formatado}: {", ".join(aulas_do_dia)}')
    else:
        print(f'Em {dia_formatado}: Nenhuma aula.')
else:
    print(f'Desculpe, o dia "{dia_desejado}" não foi encontrado no seu dicionário de aulas.')
```
