```python
nome_usuario = input("Digite o nome do usuário: ")
senha = input("Digite a senha: ")

while True:  # "True" deve estar com T maiúsculo
    if senha == nome_usuario:
        print("Erro: A senha não pode ser igual ao nome do usuário.")
        senha = input("Digite uma nova senha: ")  # pede novamente a senha
    else:
        print("Nome do usuário e senha aceitos.")
        break

print("Fim do programa.")

```
