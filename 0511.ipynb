```python
lista = []

while True:
  num = int(input("Digite um númer inteiro ( 0 para sair): "))
  if num == 0:
    break
  lista.append(num)

  soma = sum(lista)
  subtracao = max(lista) - min(lista)
  maior_num = max(lista)
  menor_num = min(lista)

print(lista)
print("a Soma é: ", soma)
print("A subtração é: ",subtracao)
print("O maior número é: ", maior_num)
print("O menor número é: ",menor_num)



```

```python
lista = []

while True:
    palavra = input("Digite uma palavra (0 para sair): ")
    if palavra == "0":
        break
    lista.append(palavra)

palavra_contar = input("Informe a palavra que você deseja contar: ")

if palavra_contar in lista:
    contar = lista.count(palavra_contar)
    print(f"A palavra '{palavra_contar}' aparece {contar} vez(es) na lista.")
else:
    print("A palavra que você escolheu não está na lista.")


```

```python
lista = []

while True:
    num = int(input("Digite um número inteiro (0 para sair): "))
    if num == 0:
        break
    lista.append(num)

sem_repeticao = list(set(lista))


repetidos = []
for n in lista:
    if lista.count(n) > 1 and n not in repetidos:
        repetidos.append(n)

print("Números digitados:", lista)
print("Valores repetidos:", repetidos)
print("Sem repetição:", sem_repeticao)

```
